# 단순연결 리스트 수업자료 응용문제 2번

```python
#로또 추첨하기 프로그램
```

```python
import random #random함수를 사용하기 위해

class Node():
  def __init__ (self): #단순 연결리스트를 사용하기 위한 초기화
    self.data = None
    self.link = None

def printNodes(start): #출력함수
  current = start
  if current == None:
    return 
  print(current.data, end = ' ')
  while current.link != None: #current의 다음 링크가 없을 때까지
    current = current.link
    print(current.data, end = ' ')
  print()

def makeLottoList(num): #로또 리스트 만들기
  global memory, head, current, pre #global 변수 선언

  node = Node()
  node.data = num
  memory.append(None)
  if head == None: #헤드가 없으면
    head = node
    return

  if head.data > num: #만약 들어온 데이터보다 헤드값이 크면
    node.link = head
    head = node #들어온 작은 값을 헤드로 설정해주기 (sort기능을 한다고 생각)
    return 

  current = head
  while current.link != None:
    pre = current
    current = current.link
    if current.data > num:
      pre.link = node
      node.link = current
      return

  current.link = node

def findNumber(num):
  global memory, head, current, pre

  if head == None:
    return False
  current = head
  if current.data == num:
    return True

  while current.link != None:
    current = current.link
    if current.data == num:
      return True
  return False

#전역 변수 선언
memory = []
head, current, pre = None, None, None

#메인 함수의 시작
if __name__ == "__main__":
  lottoCount = 0
  while True:
    lotto = random.randint(1,45) #1~45 사이의 랜덤값 부여
    if findNumber(lotto):
      continue
    lottoCount += 1 #로또 번호 생성 갯수를 카운트
    makeLottoList(lotto)
    if lottoCount >= 6: #로또 번호가 6개가 다 차면
      break #브레이크

  printNodes(head)
```

# 스택 수업자료 응용예제-2

```python
#파일 내용을 완전히 거꾸로 출력하기
#여기서는 with open("진달래꽃.txt", 'r', encoding = 'UTF8') as rfp:를 실행할 txt파일을 추가할 수 
#없으므로 결과물을 이미지를 통해 올린다
#전체 코드는 밑에와 같다
```

```python
def isStackFull(): #스택이 차있는지 판별
  global SIZE, stack, top
  if(top >= SIZE-1):
    return True
  else:
    return False

def isStackEmpty(): #스택이 비어있는지 판별
  global SIZE, stack, top
  if (top == -1):
    return True
  else:
    return False

def push(data): #스택에 데이터 집어넣기 (push)
  global SIZE, stack, top
  if(isStackFull()): #꽉차있으면
    return #리턴
  top += 1 #아니면 top을 한 층 올리고
  stack[top] = data #스택에 꼭대기에 data를 추가

def pop(): #스택에 데이터 빼기(pop)
  global SIZE, stack, top
  if(isStackEmpty()): #스택이 비어있으면
    return None #None을 반환
  data = stack[top]
  stack[top] = None #꼭대기를 None으로 만들고
  top -= 1 #top을 1층 줄인다
  return data

def peek():
  global SIZE, stack, top
  if(isStackEmpty()):
    return None
  return stack[top]

#전역 변수 선언 부분
SIZE = 100
stack = [None for _ in range(SIZE)]
top = -1

#메인함수
if __name__ =="__main__":
  with open("진달래꽃.txt", 'r', encoding = 'UTF8') as rfp:
    lineAry = rfp.readlines()

  print("----원본----")
  for line in lineAry:
    push(line)
    print(line, end = '')
  print()

  print("----거꾸로 처리된 결과----")
  while True:
    line = pop()
    if line == None:
      break

    miniStack = [None for _ in range(len(line))]
    miniTop = -1

    for ch in line:
      miniTop += 1
      miniStack[miniTop] = ch

    while True:
      if miniTop == -1:
        break
      ch = miniStack[miniTop]
      miniTop -= 1
      print(ch, end = '')
```

# 큐 수업자료 응용예제-2

```python
def isQueueFull():
  global SIZE, queue, front, rear 
  if((rear+1) % SIZE == front):
    return True
  else:
    return False

def isQueueEmpty():
  global SIZE, queue, front, rear 
  if(front == rear):
    return True
  else:
    return False

def enQueue(data):
  global SIZE, queue, front, rear 
  if(isQueueFull()):
    print("큐가 꽉 찼습니다.")
    return
  rear = (rear+1) % SIZE
  queue[rear] = data

def deQueue():
  global SIZE, queue, front, rear
  if (isQueueEmpty()):
    print("큐가 비어있습니다.")
    return None
  front = (front + 1) % SIZE
  data = queue[front]
  queue[front] = None
  return data

def peek():
  global SIZE, queue, front, rear
  if(isQueueEmpty()):
    print("큐가 비었습니다.")
    return None
  return queue[(front+1) % SIZE]

def calcTime():
  global SIZE, queue, front, rear
  timeSum = 0
  for i in range((front+1)%SIZE, (rear+1)%SIZE):
    timeSum += queue[i][1]
  return  timeSum

#전역변수
SIZE = 6
queue = [None for _ in range(SIZE)]
front = rear = 0

#메인 코드 부분
if __name__ == "__main__":
  waitCall = [('사용', 9), ('고장', 3), ('환불', 4), ('환불', 4),('고장', 3)]

  for call in waitCall:
    print("귀하의 대기 예상시간은", calcTime(),"분입니다.")
    print("현재 대기 콜 --> ", queue)
    enQueue(call)
    print()

  print("최종 대기 콜 -->", queue)
  print("프로그램 종료!")
```
